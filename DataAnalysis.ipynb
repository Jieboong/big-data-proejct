```python
import pandas as pd
import boto3
from datetime import datetime
```

```python
session = boto3.Session(profile_name='default')
s3 = session.client('s3')
bucket = '2018102232'
today = datetime.today().strftime("%Y%m%d")
```

```python
s3.download_file('2018102232','olive/{today}.csv'.format(today=today), './olive_{today}.csv'.format(today=today))
s3.download_file('2018102232','coupang/{today}.csv'.format(today=today),'./coupang_{today}.csv'.format(today=today))
```

```python
coupang_df = pd.read_csv('coupang_{today}.csv'.format(today=today))
olive_df = pd.read_csv('olive_{today}.csv'.format(today=today))
```

```python
coupang_df.drop('quantity', axis=1, inplace=True)
```

```python
coupang_df['purchase']='coupang'
olive_df['purchase'] = 'olive_young'
```

```python
merged_df = pd.concat([coupang_df, olive_df], axis=0, ignore_index=True)
```

```python
merged_df['price ver volume'] = merged_df['new_price'] / merged_df['volume'] 
```
